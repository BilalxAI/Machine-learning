```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1,2,3,4,5])
y_train=np.array([30,50,70,90,110])
m=len(x_train)
print(m)
```

```python
plt.scatter(x_train,y_train)
plt.title("predicting salaries")
plt.xlabel("years of experience")
plt.ylabel("prices in thosuands")
plt.legend()
plt.show()
```

```python
def compute_f (x,w,b):
  m=len(x_train)
  f=np.zeros(m)
  for i in range(m):
    f[i]=w*x[i]+b
  return f
```

```python
w=21.8
b=1
```

```python
f=compute_f(x_train,w,b)
plt.plot(x_train,f,c="y",label="prediction")
plt.scatter(x_train,y_train)
plt.title("predicting salaries")
plt.xlabel("years of experience")
plt.ylabel("prices in thosuands")
plt.legend()
plt.show()
```
